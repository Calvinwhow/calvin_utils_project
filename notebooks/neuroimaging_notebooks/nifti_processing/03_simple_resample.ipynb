```python
from glob import glob
import nibabel as nib
from nilearn.image import resample_img

pattern='/Volumes/HowExp/datasets/03h_ADVANCE_Alzheimer_DBS/subjects/sub-*/stimulations/MNI_ICBM_2009b_NLIN_ASYM/stimpyper/*/*.nii*'
trg = '/Users/cu135/Partners HealthCare Dropbox/Calvin Howard/studies/ccm_memory/supplement_similarity_to_optimal_vtas/mask.nii.gz'
trg = nib.load(trg)

for p in glob(pattern):
    fl = nib.load(p)
    
    fl_resampled = resample_img(
    fl,
    target_affine=trg.affine,
    target_shape=trg.shape,
    interpolation='nearest',
    force_resample=True,
    copy_header=True
    )
    p = p.strip(".gz")
    p = p.strip(".nii")
    fl_resampled.to_filename(p + "_resampled.nii.gz")
```
