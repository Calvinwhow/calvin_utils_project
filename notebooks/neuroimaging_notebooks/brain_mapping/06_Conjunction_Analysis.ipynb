# 03 - Run the Conjunction Analysis

Options for 'method' argument
- 'agreement': will keep all positive and negative regions that are shared between the two maps. Does not use thresholds, so is best used with pre-thresholded maps (such as FWE T maps).
- 'signed': will apply threshold to maps, then only keep regions that share a sign between the two maps. 
- 'unsigned': will apply threshold to maps, but keeps all regions meeting threshold across maps. 

```python
from calvin_utils.nifti_utils.conjunction_analysis import ConjunctionMap
conj = ConjunctionMap(
    nift_path_1=nifti_path_1,
    nift_path_2=nifti_path_2,
    threshold1=nifti_1_threshold,
    threshold2=nifti_2_threshold,
    direction1=nifti_1_threshold_direction,
    direction2=nifti_2_threshold_direction,
    mask_path="/Users/cu135/hires_backdrops/MNI/MNI152_T1_2mm_brain_mask.nii",
    output_dir=out_directory,
    output_name=filename,
    method='agreement'
)
img = conj.run()  # numpy array of surviving voxel indices
img
```
