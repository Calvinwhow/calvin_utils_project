```python
import subprocess
import pandas as pd

def process_rois(df):
    for index, row in df.iterrows():
        save_path = row['subject']
        for col in df.columns[1:]:  # Skip the 'subject' column
            if 'Unnamed' in col:  # Skip unnamed columns
                continue
            roi_path = col
            if pd.notna(row[col]):  # Check if the cell is not empty
                value_to_multiply = row[col]  # The value to multiply with
                output_path = f"{save_path}/{roi_path.split('/')[-1]}"
                command = f"fslmaths \"{roi_path}\" -mul {value_to_multiply} \"{output_path}\""
                subprocess.run(command, shell=True)

# process_rois(df)  # Uncomment and run this line in your Jupyter notebook
```
